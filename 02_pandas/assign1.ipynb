```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
brand_name = np.array(['Maruti Suzuki', 'Hyundai Motor', 'Tata', 'Mahindra', 'Kia', 'Toyota'])
cars_sold = [1479221, 525088, 404843, 323691, 232570, 157893]
```

```python
plt.figure(figsize=(8, 6))
plt.pie(cars_sold, labels=brand_name, autopct='%1.1f%%', startangle=140)
plt.title('Car Sales by Brand (Pie chart)')
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(brand_name, cars_sold, color='darkblue')
plt.title('Car Sales by Brand (Scatter Plot)')
plt.xlabel('Brand')
plt.ylabel('Cars Sold')
```

```python
plt.figure(figsize=(8, 6))
plt.plot(brand_name, cars_sold, marker='o', color='b', linestyle='-', linewidth=2, markersize=8)
plt.title('Car Sales by Brand (Line Graph)')
plt.xlabel('Brand')
plt.ylabel('Cars Sold')
```

```python
plt.figure(figsize=(8, 6))
plt.bar(brand_name, cars_sold, color=['skyblue', 'lightcoral', 'lightblue', 'lightgreen', 'lightpink', 'lightyellow'])
plt.title('Car Sales by Brand (Stacked Bar Chart)')
plt.xlabel('Brand')
plt.ylabel('Cars Sold')
```

```python
plt.figure(figsize=(8, 6))
plt.hist(cars_sold, bins=10, color='skyblue', edgecolor='black')
plt.title('Car Sales Histogram')
plt.xlabel('Cars Sold')
plt.ylabel('Frequency')
```

```python
df = pd.read_csv('tips.csv')
```

```python
df.columns
```

```python
# Scatter plot
plt.scatter(df['total_bill'], df['tip'])
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.title('Total Bill vs. Tip')
plt.show()
```

```python
# Histogram
plt.hist(df['total_bill'], bins=10, edgecolor='k')
plt.xlabel('Total Bill')
plt.ylabel('Frequency')
plt.title('Total Bill Distribution')
plt.show()
```

```python
# Bar chart
day_totals = df.groupby('day')['total_bill'].sum()
days = day_totals.index
totals = day_totals.values

plt.bar(days, totals)
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.title('Total Bill by Day')
plt.show()
```

```python
# Box plot
plt.boxplot([df[df['day'] == day]['total_bill'] for day in df['day'].unique()], labels=df['day'].unique())
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.title('Total Bill Distribution by Day')
plt.show()
```

```python
# Pie chart
time_counts = df['time'].value_counts()
labels = time_counts.index
sizes = time_counts.values

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.axis('equal')
plt.title('Time Distribution')
plt.show()
```

```python
df_sorted = df.sort_values(by='total_bill')

# Line graph
plt.plot(df_sorted['total_bill'], label='Total Bill')
plt.xlabel('Index')
plt.ylabel('Total Bill')
plt.title('Total Bill Over Index')
plt.legend()
plt.show()
```

```python
# Selecting the first 10 rows
df_sorted = df.sort_values(by='total_bill')

# Selecting rows 11 to 30
df_subset = df_sorted.iloc[20:30]

# Line plot
plt.plot(df_subset['total_bill'], df_subset['tip'], label='Tips')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.title('Total Bill vs. Tip (Rows 11-30)')
plt.legend()
plt.show()
```
